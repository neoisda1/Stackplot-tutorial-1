```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
sleeping = [7, 6, 4, 5, 9]
eating =[1, 2, 1, 0.5, 3]
working = [7, 4, 3, 8, 10]
playing = [1, 2, 3, 1, 4]
plt.plot([],[], color = 'red', label = 'sleeping', linewidth = 5)
plt.plot([],[], color = 'cyan', label = 'eating', linewidth = 5)         
plt.plot([],[], color = 'pink', label = 'working', linewidth = 5)
plt.plot([],[], color = 'blue', label = 'playing', linewidth = 5)
plt.stackplot(days, sleeping,eating,working,playing, colors = ['red', 'cyan', 'pink', 'blue'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting graph\ncheck it out')
plt.legend()
plt.show()
```
